```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


# ----------------------------
# CHANGE THESE TO YOUR FILES
# ----------------------------
file1 = "/content/logs_probability_0.8.xlsx"
file2 = "/content/logs_with_probability_0.65.xlsx"


def load_log(path):
    df = pd.read_excel(path)

    # Ensure numeric columns
    numeric_cols = [
        "elapsed_s",
        "alt_m",
        "x_est",
        "y_est",
        "x_err",
        "y_err",
        "vx",
        "vy",
        "vz",
        "centered",
    ]

    for col in numeric_cols:
        if col in df.columns:
            df[col] = pd.to_numeric(df[col], errors="coerce")

    return df


df1 = load_log(file1)
df2 = load_log(file2)

# Estimated Position
plt.figure()
plt.plot(df1["elapsed_s"], df1["x_est"], label="0.8 X_est")
plt.plot(df1["elapsed_s"], df1["y_est"], label="0.8 Y_est")
plt.plot(df2["elapsed_s"], df2["x_est"], "--", label="0.65 X_est")
plt.plot(df2["elapsed_s"], df2["y_est"], "--", label="0.65 Y_est")
plt.xlabel("Time (s)")
plt.ylabel("Estimated Position")
plt.title("Kalman Estimated Position")
plt.legend()
plt.grid()
plt.show()


# Image Error
plt.figure()
plt.plot(df1["elapsed_s"], df1["x_err"], label="0.8 X_err")
plt.plot(df1["elapsed_s"], df1["y_err"], label="0.8 Y_err")
plt.plot(df2["elapsed_s"], df2["x_err"], "--", label="0.65 X_err")
plt.plot(df2["elapsed_s"], df2["y_err"], "--", label="0.65 Y_err")
plt.xlabel("Time (s)")
plt.ylabel("Image Error")
plt.title("Image Error Comparison")
plt.legend()
plt.grid()
plt.show()

# Velocity Commands
plt.figure()
plt.plot(df1["elapsed_s"], df1["vx"], label="0.8 VX")
plt.plot(df1["elapsed_s"], df1["vy"], label="0.8 VY")
plt.plot(df2["elapsed_s"], df2["vx"], "--", label="0.65 VX")
plt.plot(df2["elapsed_s"], df2["vy"], "--", label="0.65 VY")
plt.xlabel("Time (s)")
plt.ylabel("Velocity (m/s)")
plt.title("Velocity Commands")
plt.legend()
plt.grid()
plt.show()


#Vertical Velocity
plt.figure()
plt.plot(df1["elapsed_s"], df1["vz"], label="0.8 VZ")
plt.plot(df2["elapsed_s"], df2["vz"], "--", label="0.65 VZ")
plt.xlabel("Time (s)")
plt.ylabel("Vertical Velocity (m/s)")
plt.title("Descent Profile")
plt.legend()
plt.grid()
plt.show()

#Altitude
plt.figure()
plt.plot(df1["elapsed_s"], df1["alt_m"], label="0.8 Alt")
plt.plot(df2["elapsed_s"], df2["alt_m"], "--", label="0.65 Alt")
plt.xlabel("Time (s)")
plt.ylabel("Altitude (m)")
plt.title("Altitude Profile")
plt.legend()
plt.grid()
plt.show()

# Centered Flag
plt.figure()
plt.plot(df1["elapsed_s"], df1["centered"], label="0.8 Centered")
plt.plot(df2["elapsed_s"], df2["centered"], "--", label="0.6 Centered")
plt.xlabel("Time (s)")
plt.ylabel("Centered (0/1)")
plt.title("Centered State")
plt.legend()
plt.grid()
plt.show()
```
